```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

## use the data from the file, read it to a dataframe and make any modification that you may find useful to solve the task

```python
ab = pd.read_csv("../data/ab-test.csv")
```

```python
after = ab[ab['time'] == 'after']
after_test = after[after['group'] == 'test']['diff'].tolist()
after_control = after[after['group'] == 'control']['diff'].tolist()
before = ab[ab['time'] == 'before']
before_test = before[before['group'] == 'test']['diff'].tolist()
before_control = before[before['group'] == 'control']['diff'].tolist()
print(after_test)
print(after_control)
print(before_test)
print(before_control)
after_lst = [after_control, after_test]
before_lst = [before_control, before_test]
```

## the figsize is still the same, you can choose whatever fontsize you like
## the color palette should be the same as in the example
## the fontsize of the title is 15
## the width of the box lines is 3, the width of the median lines is 2

```python
style_data =  {'boxprops': dict(color='g', linewidth=3),
               'capprops': dict(color='r'),
               'whiskerprops': dict(color='r'),
               'medianprops': dict(color='r', linewidth=2)
               }

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 8), sharey='all')
plt.rc('font', size=10)
fig.suptitle('Lag before labs deadlines. After and before Newsfeed', fontsize=15)
axes[0].set(title='after')
axes[0].set_xlabel('[group]')
axes[0].grid(axis='both')
axes[0].boxplot(after_lst, showmeans=True, labels=['control', 'test'], **style_data)
axes[1].set(title='before')
axes[1].set_xlabel('[group]')
axes[1].grid(axis='both')
axes[1].boxplot(before_lst, showmeans=True, labels=['control', 'test'], **style_data)
# fig =  plt.figure(figsize=(15,8))
# plt.hist(after_test, bins=24, range=[0, 24], label='working_day', color='pink')
# plt.hist(before_control, bins=24, range=[0, 24], label='weekend', alpha=0.7, color='cyan')
# plt.xlim(0,24)
# plt.legend()
# plt.show()
```
